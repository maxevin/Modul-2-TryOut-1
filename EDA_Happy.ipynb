```python
data_soc = data[['Social support']]
data_soc.sort_values(by=['Social support'], ascending=True, inplace=True)
data_soc
```

```python
fig, ax = plt.subplots()
ax.hist(data['Freedom to make life choices'], bins = 10, histtype='step')
ax.set_title('Freedom to make life Choices Histogram', size=15)
ax.set_xlabel('Satisfaction', size = 10)
ax.set_ylabel('Frequency', size = 10)
ax.grid()
```

```python
gdp_score = np.corrcoef(data['GDP per capita'],data['Score'])
gdp_score
```

```python
healthy_score = np.corrcoef(data['Healthy life expectancy'],data['Score'])
healthy_score
```

```python
above_avg = data[data['Score'] > data['Score'].mean()]
above_avg
```

```python
top5_gdp = above_avg[['Country or region','GDP per capita']]
top5_gdp.sort_values(by=['GDP per capita'],ascending=False, inplace=True)
data_gdp = top5_gdp.iloc[0:5]
data_gdp
```

```python
fig, ax = plt.subplots()
ax.bar(data_gdp['Country or region'], data_gdp['GDP per capita'], color='maroon')
ax.set_xticklabels(data_gdp['Country or region'], rotation=45)
ax.set_title('Top 5 Country with Highest GDP', size = 15)
ax.set_xlabel('Country', size = 12)
ax.set_ylabel('GDP per capita',size = 12)
ax.grid()
```

```python
top5_perce = above_avg[['Country or region','Perceptions of corruption']]
top5_perce.sort_values(by=['Perceptions of corruption'],ascending=True, inplace=True)
data_perce = top5_perce.iloc[0:5]
data_perce
```

```python
fig, ax = plt.subplots()
ax.bar(data_perce['Country or region'], data_perce['Perceptions of corruption'],color='green')
ax.set_xticklabels(data_perce['Country or region'], rotation=45)
ax.set_title('Top 5 Country with Lowest Perceptions of Corruption', size = 15)
ax.set_xlabel('Country', size = 12)
ax.set_ylabel('Perceptions of corruption',size = 12)
ax.grid()
```

```python
top5_geno = above_avg[['Country or region','Generosity']]
top5_geno.sort_values(by=['Generosity'],ascending=True, inplace=True)
data_geno = top5_geno.iloc[0:5]
data_geno
```

```python
fig, ax = plt.subplots()
ax.bar(data_geno['Country or region'], data_geno['Generosity'])
ax.set_xticklabels(data_geno['Country or region'], rotation=45)
ax.set_title('Top 5 Country with Lowest Generosity', size = 15)
ax.set_xlabel('Country', size = 12)
ax.set_ylabel('Generosity',size = 12)
ax.grid()
```
